# Yandex AI Studio — проверка подключения

Перед запуском: `cp .env.example .env` и вставь ключи.

```python
from dotenv import load_dotenv
import os

load_dotenv("../.env")

folder_id = os.environ.get("YANDEX_FOLDER_ID", "")
api_key = os.environ.get("YANDEX_API_KEY", "")
model_name = os.environ.get("YANDEX_MODEL", "yandexgpt-lite")

if not folder_id or not api_key:
    raise ValueError("Заполни YANDEX_FOLDER_ID и YANDEX_API_KEY в .env")

print("ok, ключи на месте")
```

```python
from yandex_ai_studio_sdk import AIStudio

sdk = AIStudio(folder_id=folder_id, auth=api_key)
print("sdk ready")
```

```python
model = sdk.models.completions(model_name).configure(temperature=0.3)
result = model.run("Скажи одним предложением, что такое электроэкстракция никеля.")

for alt in result:
    print(alt)
```

```python
emb = sdk.models.text_embeddings("text-search-doc")
vector = emb.run("электроэкстракция никеля")

print(len(vector))
print(vector.num_tokens)
```
